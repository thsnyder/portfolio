# Violence Against Women and Girls
This dataset aggregates agreement with key questions across gender, education level and many other socioeconomic variables from 70 different countries.
The data was collected as part of the Demographic and Health Surveys (DHS) program, which exists to advance the global understanding of health and population trends in developing countries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('violence_data.csv')
df.head()
```

```python
Zambia = df[df.Country == 'Zambia']
Zambia.head(3)
```

```python
# graph = Zambia[Zambia['Demographics Question'] == 'Education']
graph = df[df['Demographics Question'] == 'Education']

g = sns.catplot(x='Demographics Response', y='Value', hue='Question', 
                order=['Primary', 'Secondary', 'Higher'], 
                data=graph, kind='bar', ci=None, height=6, aspect=1.5)
g.set_axis_labels('Education Level', 'Percentage(%)')
g.fig.suptitle('Those agreeing a husband is justified in hitting his wife', size=15, y=1.05)
# plt.savefig('VAWGcountries_Education_plot.png')
```

```python
Rwanda = df[df.Country == 'Rwanda']
graph = Rwanda[(Rwanda['Demographics Question'] == 'Residence') & (Rwanda['Gender'] == 'F')]

g = sns.catplot(x='Demographics Response',y='Value',
                order=['Rural','Urban'],
                data=graph,kind='bar',ci=None)
g.set_axis_labels('Residence Type','Percentage (%)')
g.fig.suptitle('Rwandan women agreeing a husband is justified in hitting his wife',y=1.05)
```

```python
Afghanistan = df[df.Country == 'Afghanistan']
graph = Afghanistan[(Afghanistan['Demographics Question'] == 'Age') & (Afghanistan['Gender'] == 'F')]

g = sns.catplot(x='Demographics Response',y='Value',
                data=graph,kind='bar',ci=None)
g.set_axis_labels('Age', 'Percentage(%)')
g.fig.suptitle('Afghan women agreeing a husband is justified in hitting his wife',y=1.05)
# plt.savefig('VAWGcountries_WomenAges_plot.png')
```

```python
cts = df[(df.Country == 'India') | (df.Country == 'Ethiopia')]
cts
```

```python
x = cts[(cts.Country == 'India') & (cts['Demographics Question'] == 'Age')]

plt.bar(x['Demographics Response'], x.Value, color='rgb',align='center')
plt.xlabel('Age')
plt.ylabel('Percentage')
plt.title('Indians agreeing a husband is justified in hitting his wife', size=15)
plt.show()
```

```python
graph = cts[cts['Demographics Question'] == 'Education']

g = sns.catplot(x='Demographics Response', y='Value', col='Country', row='Gender', hue='Question', 
                order=['Primary', 'Secondary', 'Higher'], 
                data=graph, kind='bar', ci=None)
g.set_axis_labels('Education Level', 'Percentage(%)')
g.fig.suptitle('Male/Female agreeing a husband is justified in hitting his wife',size=16, x=.4, y=1.05)
plt.savefig('VAWGcountries_Education_plot.png')
```

```python
graph = cts[(cts['Demographics Question'] == 'Age') & (cts['Gender'] == 'F')]

g = sns.catplot(x='Demographics Response',y='Value',
                data=graph,kind='bar',ci=None)
g.set_axis_labels('Age', 'Percentage(%)')
g.fig.suptitle('Those agreeing a husband is justified in hitting his wife',y=1.05)
```

```python
means = df.groupby('Country').mean('Value')
means_sorted = means.sort_values('Value',ascending=False)
means_sorted.reset_index(inplace=True)
means_sorted = means_sorted.rename(columns = {'index':'Country'})
x = means_sorted.Value
y = means_sorted.Country

fig, g = plt.subplots(figsize = (20,30))
sns.set_context("paper", font_scale=2, rc={"lines.linewidth": 5})

g = sns.barplot(x, y, orient='h', alpha=1)
g.set_xlabel('Percentage(%)', fontsize=20)
g.set_ylabel('Country', fontsize=20)
g.xaxis.set_minor_locator(plt.MultipleLocator(1))
g.xaxis.set_major_locator(plt.MultipleLocator(5))
g.xaxis.grid(True)
g.set_title('Those agreeing a husband is justified in hitting his wife', 
            fontdict= { 'fontsize': 26, 'fontweight':'bold'})

# plt.savefig('VAWGcountriesB_plot.png')
```
